```python
#Se crea un metodo para limpiar y castear los datos que vienen en formato string de la columna height para convertilo en float y asi hallar la media
def cast_height_column(value):
    try:
        # Busca un número seguido de 'cm'
        return float(value.lower().split('cm')[0])
    except (ValueError, AttributeError, IndexError):
        return np.nan  # Si no se pone un NaN en su lugar
```

```python
# Metodos para calcular promedio, mediana, desviación estándar y percentiles
def calcular_promedio(df,cols, posicion, anio):
    for col in cols:
        if col in df.columns and df[col].notna().any():
            promedio = np.nanmean(df[col])
            print(f"El promedio para el atributo {col} de la posicion {posicion} del año {anio} es :{np.round(promedio, 2)}")
def calcular_estadisticas(df, cols,posicion, anio):
    for col in cols:
        try:
            if col in df.columns and df[col].notna().any():
                # Dejaremos el parámetro overwrite_input a False ya que no es necesario para calcular la mediana
                # Este parámetro  ahorra memoria modificando el array original, lo cual no es relevante aquí
                mediana = np.nanmedian(df[col], overwrite_input=False)
                print(f"La mediana para el atributo {col} de la posicion {posicion} del año {anio} es :{np.round(mediana, 2)}")
                #Los datos tan dispersos en el calculo de las desviaciones del ovr y del pac se deben a los outliers anteriormente mencionados pues a estar
                #tan alejados de la mediana disparan la desviacion
                desviacion = np.nanstd(df[col])
                print(f"La desviación estándar para el atributo {col} de la posicion {posicion} del año {anio} es : {np.round(desviacion, 2)}")
                # Ahora calcularemos los percentiles en este caso calcularemos Q1,Q2 y Q3 que es donde caen el 25%, 50% y 75 % de los datos
                q1 = np.nanpercentile(df[col], 25)
                q2 = np.nanpercentile(df[col], 50)
                q3 = np.nanpercentile(df[col], 75)
                print(f"Los Percentiles (Q1, Q2, Q3) para el atributo {col} de la posicion {posicion} del año {anio} es : {np.round(q1, 2)}, {np.round(q2, 2)}, {np.round(q3, 2)}")
        except KeyError as ex:
            print(f"No existe la clave {ex} para la posicion {posicion} del año {anio}")
```

#### Calcular estadísticas descriptivas

```python
for anio, positions in df_players.items():
    for pos, df in positions.items():
        print(f"\nEstadísticas para {pos} en {anio}:")
        try:
            if df_players[anio][pos]['height'].str.contains('cm', case=False, na=False).any():
                df_players[anio][pos].loc[:, 'height'] = df_players[anio][pos]['height'].apply(cast_height_column)
        except (AttributeError, TypeError):
            pass
        calcular_promedio(df,['ovr','pac','age','height'], pos, anio)
        calcular_estadisticas(df,['ovr','pac'], pos, anio)
```

```python
# Y ahora realizamos la prueba de distribucion normal, y sacamos en conclusion
#Para nuestra prueba y teniendo en cuenta  que nuestro dataset es demasiado grande realizaremos la prueba de normalidad en lugar de 
# Con Shapiro-Wilk que trabaja con numeros pequeños (no mas de 5000 registros) probaremos con Kolmogorov-Smirnov Test o K-S Test 
stat, p = kstest(ovr_all['ovr'], 'norm', args=(ovr_all['ovr'].mean(), ovr_all['ovr'].std()))
print("p-valor:", p)
```
